## Variables and Strings
We define a variable 'text-sample' and assign ita string of Yorùbá texts

```python
text_sample = "Open Arts is proud to present the Yoruba  translation of Mr. Walser,a short story by Gonçalo M. Tavares, rendered into our indigenous language as part of our modest but determined contribution to literary development in Nigeria.This publication marks the ourth title from our young but purposeful publishing outfit based in Kaduna."

print("Source text: " )
print(text_sample)
print("------")
```

## Lists and Basic Normalization
The `.split()` method breaks the string at every space, returning a list of words.

```python
word_list = text_sample.split()

print("Tokenized  Word List: ")
print(word_list)
print("Total word count:", len(word_list))
print("--------")
```

## Dictionary and Loop


```python
# We creat an empty Dictionary bucket to hold our character frequency. Frequency means "The appearance of each word".

character_frequency = {}

# Then, we loop through every character in the "text_sample". At the same time, we use .lower() to ensure "Ẹ" and "ẹ" are counted as a single character.

for character in text_sample.lower():
    # we only want to count the letter, ignoring spaces and puntuactions like "?", "," and "."
    # python's .isalpha() method correctly identifies standard letters and those with tonal diacritics.
    if character.isalpha():
        # if the character is already in our dictionary add 1 to the count
        if character in character_frequency:
            character_frequency[character] = character_frequency[character] + 1

            # if we have not seen this character yet, add it to the dictionary bucket with a count of 1
        else:
            character_frequency[character] = 1

print("Character Frequency Distribution: ")
print(character_frequency)
```
