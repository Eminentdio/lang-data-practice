```python
import re

messy_data = """ Please, send the trabscribed corpus data to data.team@linguistics.edu or adun123@regez.com by 15/08/2026.
Ensure that words with tonal marks like ọ̀pákùtẹ̀lẹ̀, ìgbéyàwó, and fẹ́ra are encoded correctly.
If there are error, contact the QA lead at qa_admin@institute.org before 12/10/2026.
"""

print("Original Text Loaded.\n---")
```

```python
# Extracting email

# Pattern = [word character/dot/dash] + @ + [word character/dot/dash] + . + letter

email_pattern = r"[\w.-]+@[\w.-]+\.[a-zA-Z0-9_]{2,}"
found_emails = re.findall(email_pattern, messy_data)

print("Extracted Emails: ")
print(found_emails)
print("------")
```

```python
# Extracting date

# Patterns = exactly 2 digits, a slash, 2 digits, a slash, 4 digits

date_pattern = r"\d{2}/\d{2}/\d{4}"
found_dates = re.findall(date_pattern, messy_data)

print("Extracted Date: ")
print(found_dates)
print("-----")
```

```python
# Extracting tonal words characters in the text
# Pattern = word boundary (\b), any word (\w*) +[our specific character] + any word (\w*), word boundary (\b)

tonal_pattern = r"\b\w*[áàẹọṣíìúùóò]\w*\b"
found_tones = re.findall(tonal_pattern, messy_data)

print(found_tones)
```
